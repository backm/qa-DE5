```python
#code for loading .csv files in
import pandas as pd

books_df = pd.read_csv("Data/03_Library Systembook.csv")
customers_df = pd.read_csv("Data/03_Library SystemCustomers.csv")

customers_df.head()
```

```python
books_df.head(25)
```

```python
books_df["Book checkout"] = pd.to_datetime(
    books_df["Book checkout"],
    format="%d/%m/%Y",
    errors="coerce"
)

books_df["Book Returned"] = pd.to_datetime(
    books_df["Book Returned"],
    format="%d/%m/%Y",
    errors="coerce"
)
```

```python
books_df["Book checkout"].head()
```

```python
books_df["Book checkout"] = books_df["Book checkout"].str.replace('"', '', regex=False)
books_df["Book Returned"] = books_df["Book Returned"].str.replace('"', '', regex=False)
```

```python
date_columns = ["Book checkout", "Book Returned"]

for col in date_columns:
    # use string but remove quotes and any spaces
    raw = books_df[col].astype(str).str.replace('"', '', regex=False).str.strip()
    books_df[col + "_raw"] = raw  # keep original for audit
    
    # convert to date (UK)
    books_df[col] = pd.to_datetime(raw, format="%d/%m/%Y", errors="coerce")
```

```python
bad_checkout = books_df[books_df["Book checkout"].isna()][["Id", "Book checkout_raw"]]
bad_returned = books_df[books_df["Book Returned"].isna()][["Id", "Book Returned_raw"]]

bad_checkout, bad_returned
```
