```python
import pandas as pd

```

```python
print("Starting Library data cleaning process...")

books_file = "Data/03_Library Systembook.csv"
customers_file = "Data/03_Library SystemCustomers.csv"

books_df = pd.read_csv(books_file)
customers_df = pd.read_csv(customers_file)

print("Files loaded successfully.")
print("Books rows:", len(books_df))
print("Customers rows:", len(customers_df))

    # START CLEANING DATES

print("Cleaning date columns...")

# Remove quotes and trim any spaces
books_df["Book checkout"] = books_df["Book checkout"].astype(str).str.replace('"', '').str.strip()
books_df["Book Returned"] = books_df["Book Returned"].astype(str).str.replace('"', '').str.strip()

# Convert to date (UK)
books_df["Book checkout"] = pd.to_datetime(
    books_df["Book checkout"],
    format="%d/%m/%Y",
    errors="coerce"
)

books_df["Book Returned"] = pd.to_datetime(
    books_df["Book Returned"],
    format="%d/%m/%Y",
    errors="coerce"
)

print("Date conversion complete.")

    # CLEAN DAYS TO BORROW (FROM WEEKS)

print("Converting borrow 'weeks' to  days...")


books_df["Allowed Days"] = (
    books_df["Days allowed to borrow"]
    .str.replace(" weeks", "")
    .str.replace(" week", "")
    .astype(float) * 7
)

print("Borrow time converted.")

print("Checking for data issues...")

books_df["Invalid Checkout Date"] = books_df["Book checkout"].isna()
books_df["Invalid Return Date"] = books_df["Book Returned"].isna()

books_df["Checkout After Return"] = (
    books_df["Book checkout"] > books_df["Book Returned"]
)

# dataframe for bad rows
bad_rows = books_df[
    (books_df["Invalid Checkout Date"]) |
    (books_df["Invalid Return Date"]) |
    (books_df["Checkout After Return"])
]

print("Number of problematic rows:", len(bad_rows))


#drop bad rows from the data

clean_books_df = books_df.drop(bad_rows.index)

print("Clean rows remaining:", len(clean_books_df))

# create clean files from cleaned df

clean_books_df.to_csv("Data/books_cleaned.csv", index=False)
bad_rows.to_csv("Data/books_problem_rows.csv", index=False)

customers_df.to_csv("Data/customers_cleaned.csv", index=False)

print("Cleaning complete.")
print("Cleaned files saved.")
```
